## Principal Component Analysis

```python
print('Logistic Regression')
plt.plot(range(len(results[results['name'] == 'pca_values_lr']['MRR'])), results[results['name'] == 'pca_values_lr']['MRR'])
plt.xlabel('PCA components')
plt.ylabel('Mean Reciprocal Rank')
plt.show()
print(f'Components: {int(results[results["name"] == "pca_values_lr"].sort_values("MRR", ascending=False).head(1)["pca"])}')
print(f'MRR: {results[results["name"] == "pca_values_lr"]["MRR"].max()}')
```
